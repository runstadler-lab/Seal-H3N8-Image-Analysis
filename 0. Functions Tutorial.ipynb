```python
from rgprocessing import loadshow_img, showchannel, threshold_image, entropy_image, find_overlap, quantify_overlap
from skimage.io import imshow
```

```python
# All of the functions will be done on a single image.
file = open('HH3_trachea_111313_5.png', 'r')
```

```python
img = loadshow_img(file)
```

```python
# Show the red channel of the positive control image.
red_channel = showchannel(img, 0)
```

```python
# Show the green channel of the positive control image.
green_channel = showchannel(img, 1)
```

```python
# Show the green channel of the positive control image - there should be none.
blue_channel = showchannel(img, 2)
```

```python
red_thresholded = threshold_image(red_channel)
```

```python
# Compute the entropy of the thresholded red channel within a 10 pixel radius.
red_entropy_10 = entropy_image(red_thresholded[0], 10)
```

```python
# Compute the entropy of the thresholded red channel within a 25 pixel radius.
red_entropy_25 = entropy_image(red_thresholded[0], 25)
```

Below, the entire procedure is repeated with the Green channel.

```python
# Below, we delineate the boundaries of the red region by thresholding the entropy calculations.
red_entropy_25_thresholded = threshold_image(red_entropy_25)
```

```python
green_thresholded = threshold_image(green_channel)
```

```python
green_entropy_25 = entropy_image(green_thresholded[0], 25)
```

```python
green_entropy_25_thresholded = threshold_image(green_entropy_25)
```

```python
# Recap: Let's see the red_entropy_25_thresholded as well.
imshow(red_entropy_25_thresholded[0])
```

```python
# For the final calculation, let's compute the total number of pixels overlapping between the two red and green channel entropies.
find_overlap(red_entropy_25_thresholded[0], green_entropy_25_thresholded[0])
# Note: because the function threshold_image returns a tuple, we have to access the first element in the tuple to get the image.
```
