```python
from rgprocessing import *
from skimage.io import imshow, imsave

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# All of the functions will be done on a single image.
file = open('Seal/Ferret Trachea/SH3_fer_trachea_100814_4.tif', 'r')
```

```python
img = loadshow_img(file)
```

```python
# Show the red channel of the positive control image.
red_channel = showchannel(img, 0)
```

```python
# Show the green channel of the positive control image.
green_channel = showchannel(img, 1)
```

```python
# Show the green channel of the positive control image - there should be none.
blue_channel = showchannel(img, 2)
```

```python
red_thresholded = threshold_image(red_channel)
```

```python
red_adaptive_thresholded = adapative_threshold(red_channel, block_size=500)
```

```python
# Compute the entropy of the thresholded red channel within a 10 pixel radius.
red_entropy_10 = entropy_image(red_adaptive_thresholded, 10)
```

```python
# Compute the entropy of the thresholded red channel within a 25 pixel radius.
red_entropy_25 = entropy_image(red_adaptive_thresholded, 25)
```

```python
# Below, we delineate the boundaries of the red region by thresholding the entropy calculations.
red_entropy_25_thresholded = threshold_image(red_entropy_25)
```

```python
# Find the largetst contiguous region present in the image - this is the most contiguous region of red entropy, 
# which delineates the tracheal tissue. Note that this only has to be done for tracheal tissue, not for lung.
tracheal_tissue = large_contiguous_regions(red_entropy_25_thresholded[0], 0.1)
imshow(tracheal_tissue)
```

Below, the entire procedure is repeated with the Green channel.

```python
green_thresholded = threshold_image(green_channel, 63.000000)
```

```python
green_entropy_25 = entropy_image(green_thresholded[0], 25)
```

```python
green_entropy_25_thresholded = threshold_image(green_entropy_25, 0.261965)
```

```python
tracheal_stain = large_contiguous_regions(green_entropy_25_thresholded[0], 0.01)
imshow(tracheal_stain)
```

```python
# Recap: Let's see the red_entropy_25_thresholded as well.
imshow(tracheal_tissue)
```

```python
# For the final calculation, let's compute the total number of pixels overlapping between the two red and green channel entropies.
overlap = find_overlap(tracheal_tissue, tracheal_stain)
save_overlap('overlap_test.png', overlap * 1.0)
# Note: because the function threshold_image returns a tuple, we have to access the first element in the tuple to get the image.
```
