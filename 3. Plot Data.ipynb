```python
import pandas as pd
from math import sqrt
```

```python
lung_data = pd.read_csv('Red-Green Overlap lung.csv', index_col=0)
# Exclude Neg_lung_111313_4.png because it is an outliner.
lung_data = lung_data.drop(6, in_place=True)
lung_data
```

```python
# Get the average of lung data
lung_data_averaged = lung_data.groupby('Sample or Control').mean()
lung_data_averaged
```

```python
# Get the standard deviation of lung data
lung_data_stdev = lung_data.groupby('Sample or Control').std()
lung_data_stdev
```

```python
# Compute standard error of mean
lung_data_sem = lung_data.groupby('Sample or Control').std()/sqrt(len(lung_data.groupby('Sample or Control')))
lung_data_sem
```

```python
# Combine the data together
lung = pd.DataFrame()
lung['mean'] = lung_data_averaged['Red-Green Overlap']
lung['sem'] = lung_data_sem['Red-Green Overlap']
lung
```
